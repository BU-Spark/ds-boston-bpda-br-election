```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sum_party_votes = pd.read_csv('election data/sum_votes_US_JPN.csv')
sum_party_votes = sum_party_votes.rename({'Unnamed: 0': 'Year'}, axis='columns')
print(sum_party_votes)
```

```python
# Function to graph votes for any party
def graphPartyVotes(df, party):
    x = sum_party_votes.Year.values
    oldList = sum_party_votes[party].values
    y =[]
    for i in range(len(oldList) - 1, -1, -1):
        votes = oldList[i]
        if np.isnan(votes) == False:
            y.insert(0,votes)
        else:
            x = np.delete(x, i)

    m, b = np.polyfit(x, y, 1)
    plt.plot(x, y, 'o')
    plt.plot(x, m*x + b)
    plt.xticks(x, x)
    plt.ylim(0, 150000)
    plt.xlabel("Year")
    plt.ylabel("Votes")
    plt.title("Votes for " + party)
    plt.show()
```

```python
# Graph votes for each party
parties = sum_party_votes.columns[1:]
for party in parties:
    graphPartyVotes(sum_party_votes, party)
```
